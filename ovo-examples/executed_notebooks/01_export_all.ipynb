```python
from ovo import db, design_logic, descriptor_logic, import_export_logic
import pandas as pd
```

```python
project = db.Project.get(name='OVO Publication Examples 1')
round_ids = db.Round.select_values("id", project_id=project.id)
pool_ids = db.Pool.select_values("id", round_id__in=round_ids)
print(len(pool_ids), 'pools', pool_ids)
```

```python
project.id
```

```python
with pd.option_context('display.max_columns', 100):
    jobs = design_logic.get_design_jobs_table(project.id)
    del jobs[('Job', 'duration')]
    display(jobs)
```

```python
jobs.to_csv('../../data/results/ovo_publication_examples_1_jobs.csv')
```

```python
designs = db.Design.select_dataframe(pool_id__in=pool_ids)
designs.head()
```

```python
descriptors = descriptor_logic.get_wide_descriptor_table(pool_ids=pool_ids)
descriptors = designs.join(descriptors)
descriptors.head()
```
