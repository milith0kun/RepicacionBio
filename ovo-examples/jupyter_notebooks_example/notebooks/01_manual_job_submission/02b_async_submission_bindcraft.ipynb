## Submit workflow

```python
input_dict = {
    "design_path": "output",
    "starting_pdb": "target.pdb",
    "binder_name": "design",

    "chains": 'E',
    "target_hotspot_residues": 'E64,E88,E96',
    "lengths": [50,60],
    "number_of_final_designs": 1
}
```

```python
with open("../../data/inputs/bindcraft_input1.json", "w") as outfile:
    outfile.write(json.dumps(input_dict, indent=4))
```

```python
job_id = scheduler.submit(
    "bindcraft",
    params={
        "time_limit_seconds": 6 * 3600,
        "input_pdb": os.path.abspath("../../data/inputs/4ZXB_E6-155.pdb"), # Using trimmed structure
        "input_json_path": os.path.abspath("../../data/inputs/bindcraft_input1.json"),
        "settings_advanced": os.path.abspath("../../data/bindcraft/default_4stage_multimer.json"),
        "settings_filters": os.path.abspath("../../data/bindcraft/default_filters.json"),
    },
    submission_args=dict(
        # optional args for scheduler 
    )
)
```

```python
print(job_id)
```

```python
print(scheduler.get_status_label(job_id))
```

```python
print(scheduler.get_result(job_id))
```

```python
result = scheduler.wait(job_id)

if not result:
    print(scheduler.get_log(job_id))
    assert False, 'Job failed'
```
